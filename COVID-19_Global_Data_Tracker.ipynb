This notebook analyzes COVID-19 data from around the world to track trends in cases, deaths, and vaccinations.  
We use data from **Our World in Data** and visualize key trends across selected countries.


```python
import pandas as pd
import os

url = "https://covid.ourworldindata.org/data/owid-covid-data.csv"
filename = "owid-covid-data.csv"

# Download the file if it's not already present
if not os.path.exists(filename):
    import urllib.request
    urllib.request.urlretrieve(url, filename)

# Load the dataset
df = pd.read_csv(filename)

# Display the first few rows
df.head()
```

```python
# Check column names
print(df.columns)

# Check for missing values
df.isnull().sum()

# Dataset summary
df.describe()
```

```python
# Focus on 3 countries
countries = ['Kenya', 'United States', 'India']
df = df[df['location'].isin(countries)]

# Convert 'date' to datetime
df['date'] = pd.to_datetime(df['date'])

# Drop rows with missing critical values
df = df.dropna(subset=['total_cases', 'total_deaths'])

# Fill other missing values with 0
df = df.fillna(0)

df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
for country in countries:
    country_data = df[df['location'] == country]
    plt.plot(country_data['date'], country_data['total_cases'], label=country)

plt.title("Total COVID-19 Cases Over Time")
plt.xlabel("Date")
plt.ylabel("Total Cases")
plt.legend()
plt.grid(True)
plt.show()
```

```python
df['death_rate'] = df['total_deaths'] / df['total_cases']
```

```python
plt.figure(figsize=(10,6))
for country in countries:
    country_data = df[df['location'] == country]
    plt.plot(country_data['date'], country_data['total_vaccinations'], label=country)

plt.title("Total Vaccinations Over Time")
plt.xlabel("Date")
plt.ylabel("Total Vaccinations")
plt.legend()
plt.grid(True)
plt.show()
```

## Key Insights

- The United States had the highest total case count.
- India showed rapid vaccination progress in 2021.
- Kenya maintained a lower death rate relative to its case count.

## Observations

- Peaks in daily cases match known global variant waves.
- Vaccination surges correlate with a drop in daily new deaths.
